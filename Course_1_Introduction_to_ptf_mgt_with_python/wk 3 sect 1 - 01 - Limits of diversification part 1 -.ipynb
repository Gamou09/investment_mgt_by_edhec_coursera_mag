## Limit of diversification - part 1

(including some fancy stuff on pandas and the anatomy of market cap weighted portfolio)

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import pandas as pd
import edhec_risk_kit as erk
```

```python
ind_return = erk.get_ind_returns()
ind_return.head()
```

```python
ind_size = erk.get_ind_size()
ind_size.head()
```

```python
ind_nfirms = erk.get_ind_nfirms()
ind_nfirms.head()
```

```python
ind_nfirms.shape, ind_size.shape, ind_return.shape
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis="columns")
```

```python
total_mktcap.plot(figsize=(15,6))
```

```python
total_mktcap = ind_mktcap.sum(axis="columns")
total_mktcap
```

```python
# can't use basic division as not the same shape
ind_capweight = ind_mktcap.divide(total_mktcap, axis = 'rows')
ind_capweight.head()
```

```python
ind_capweight["2012"].sum(axis = 'columns')
```

```python
ind_capweight[["Fin", "Steel", "Food"]].plot(figsize=(12,6))
```

```python
# create a cap-weighted market index like S&P500
total_market_return = (ind_capweight * ind_return).sum(axis="columns")
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(12,6), title = "Total Market CapWeighted Index 1926-2018")
```

```python
total_market_index["1980":].plot(figsize=(12,6))
```

```python
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6))
```

## Limit of diversification - part 2

(including some fancy stuff on pandas and the anatomy of market cap weighted portfolio)

```python
total_market_index["1980":].plot(figsize=(12,6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6)) # moving average
```

## Rolling correlation - along with MultiIndexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
# example of MultiIndexes
```

```python
ind_return.head()
# simple index
```

```python
ts_corr.index.names = ["date", "industry"]
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets", figsize=(12,6), legend = True)
ind_tr36corr.plot(label="Tr36 Mo Corr", legend = True)
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets", figsize=(12,6), legend = True)
ind_tr36corr.plot(label="Tr36 Mo Corr", legend = True, secondary_y = True)
```

```python
tmi_tr36rets["2007":].plot(label="Tr36 Mo Rets", figsize=(12,6), legend = True)
ind_tr36corr["2007":].plot(label="Tr36 Mo Corr", legend = True, secondary_y = True)
```

```python
tmi_tr36rets.corr(ind_tr36corr)
# correlation between average correlation and return --> negativ mean when market tanks correlation increase 
# and can't bail you out
```
