```python
Liste_ind = list(ind49_rets.columns)
Liste_beta = []
for ind in Liste_ind:
    ind49 = np.array(ind49_rets.loc["1991":"1993"][ind].tolist()) - np.array(fff.loc["1991":"1993"]["RF"].tolist())
    mkt_excess = fff.loc["1991":"1993", ["Mkt-RF"]]
    exp_var = mkt_excess.copy()
    result = erk.regress(ind49, exp_var)
    Liste_beta.append(result.params[0])
Final_result = pd.DataFrame({
    "Industry": Liste_ind,
    "Beta": Liste_beta
})
Final_result.sort_values("Beta", ascending=False).head(),Final_result.sort_values("Beta", ascending=False).tail()
```

```python
# from question 7
fff_1991_2018 = fff.loc["1991":"2018",]
ind49_1991_2018 = ind49_rets.loc["1991":"2018",]
Liste_ind = list(ind49_rets.columns)
Liste_tilt_smb = []
for ind in Liste_ind:
    ind49_excess = np.array(ind49_1991_2018[ind].tolist()) - np.array(fff_1991_2018["RF"].tolist())
    fff_1991_2018_excess = fff_1991_2018[["Mkt-RF", "SMB", "HML"]]
    exp_var = fff_1991_2018_excess.copy()
    result = erk.regress(ind49_excess, exp_var)
    Liste_tilt_smb.append(result.params[1])
Final_result = pd.DataFrame({
    "Industry": Liste_ind,
    "Tilt SMB": Liste_tilt_smb
})
Final_result.sort_values("Tilt SMB", ascending=False).head(),Final_result.sort_values("Tilt SMB", ascending=False).tail()
```
